# Mundo 1 - O que é o Matplotlib?

### O matplotlib é a biblioteca padrão de gráficos no Python.

Mesmo que algum pacote seja utilizado (seaborn, quantstats, qualquer um que seja), esse pacote utiliza o matplotlib como base.
<br>

### O matplotlib trabalha com duas dimensões para criar os gráficos:

* Figuras
* Eixos

A partir dessas dimensões, você consegue personalizar o gráfico da cor que você quiser.

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

ax.grid(True)

plt.show()
```

### Existe inúmeros submódulos dentro do matplotlib.

* Pyplot, que cuida da plotagem dos gráficos no geral.
* Animation, que faz animações.
* Axes, que configura no detalhe os eixos.

Nessa galáxia nós iremos trabalhar com alguns (os principais). São 54 no total e muitos, como legend, são submódulos para especificações que podem ser trabalhadas dentro do Pytplot, que é o principal.

### PLT x FIG, AX:

Existem duas formas de plotar gráficos no matplotlib:

* Orientado a objeto (fig, ax)
* Estilo Pyplot (plt.plot())

A primeira forma, que já foi mostrada na aula, é a mais recomendada e tratando o gráfico como um objeto, a customização é praticamente infinita.

A segunda forma, com plt.plot, é mais simples e pode ser muito mais rápida de vizualizar. Não precisa criar um objeto bonitinho, etc.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'x': np.random.random(5),
                   'y': np.random.random(5)})

df
```

```python
#forma OO

fig, ax = plt.subplots()

ax.plot(df.index, df['x'])
ax.plot(df.index, df['y'])
ax.grid(True)
```

```python
#forma pyplot: figuras e eixos são criados automaticamente

plt.plot(df)
```
